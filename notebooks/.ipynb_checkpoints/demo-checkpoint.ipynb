```python
device = 'cpu'
```

```python
from pyro.infer import MCMC, NUTS
import pyro
import torch
```

```python
torch.__version__
```

```python
pyro.__version__
```

```python
target = torch.distributions.Normal(loc=torch.tensor([10., 10.], device=device, dtype=torch.float32),
                                                          scale=torch.tensor([1., 1.], device=device, dtype=torch.float32))
```

```python
def potential_fn(samples):
    return -target.log_prob(samples['points']).sum(-1)
```

```python
nuts = NUTS(potential_fn=potential_fn)
```

```python
num_chains = 12
initial_params = {'points': torch.randn((num_chains, 2), dtype=torch.float32, device=device)}
```

```python
mcmc = MCMC(kernel=nuts, num_samples=100, warmup_steps=100,
            initial_params=initial_params, disable_progbar=False, num_chains=num_chains, )
```

```python
mcmc.run()
```

```python
results = mcmc.get_samples(group_by_chain=True)['points']
```

```python
results.shape # n_chains, num_samples, dim
```

```python
results
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
target_samples = target.sample((1000, ))
```

```python
chain_idx = np.random.choice(np.arange(num_chains), size=(2, ))
```

```python
plt.close()
plt.figure(figsize=(6, 4), dpi=150)
plt.scatter(target_samples[:, 0], target_samples[:, 1], color='blue')
colors = {0: 'red', 1: 'green'}
for i, chain_id in enumerate(chain_idx):
    plt.plot(results[chain_id, :, 0], results[chain_id, :, 1], '--o', color=colors[i])
plt.tight_layout()
plt.show();
```
